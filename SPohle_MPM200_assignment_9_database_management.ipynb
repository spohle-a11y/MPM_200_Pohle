## Detail Vertical Join of Mammals and Birds

### Subtask:
Explain the process of vertically joining the `mammals_meta` and `birds_meta` dataframes to create a unified `metadata` dataframe. This includes identifying common columns and the method for concatenation.


**Reasoning**:
To prepare for the vertical join, I will first identify and print the common column names between `birds_species_only` and `mammals_meta` to understand which fields will be preserved in the unified `metadata` dataframe.



```python
common_cols = list(set(birds_species_only.columns) & set(mammals_meta.columns))
print(f"Common columns between birds_species_only and mammals_meta: {common_cols}")
```

**Reasoning**:
Now that common columns are identified, I will select these columns from both `birds_species_only` and `mammals_meta` to create new dataframes, and then vertically concatenate them to form the `metadata` dataframe as instructed.



```python
birds_common_cols = birds_species_only[common_cols]
mammals_common_cols = mammals_meta[common_cols]

metadata = pd.concat([birds_common_cols, mammals_common_cols], ignore_index=True)

print("First few rows of the unified metadata dataframe:")
display(metadata.head())
print(f"Shape of the unified metadata dataframe: {metadata.shape}")
```

### Vertical Join of Mammals and Birds Dataframes

To create a unified `metadata` dataframe from `birds_meta` and `mammals_meta`, we followed these steps:

1.  **Identify Common Columns**: We first determined the intersection of column names between the `birds_species_only` (which is a filtered version of `birds_meta` containing only species-level entries) and `mammals_meta` dataframes. This step ensures that only relevant and comparable fields are retained during the concatenation.

    The common columns identified were: `['Synonyms', 'Family', 'Order', 'Authority']`.

2.  **Select Common Columns**: From both `birds_species_only` and `mammals_meta`, we selected only these common columns to create two temporary dataframes: `birds_common_cols` and `mammals_common_cols`.

3.  **Vertical Concatenation**: These two temporary dataframes (`birds_common_cols` and `mammals_common_cols`) were then vertically concatenated using `pd.concat()`. The `ignore_index=True` parameter was used to reset the index of the resulting `metadata` dataframe, ensuring a continuous index across all rows from both original dataframes.

    The resulting `metadata` dataframe contains all rows from both the bird and mammal species datasets, but only for the columns that were common to both. This creates a consolidated table of taxonomic metadata for both groups, ready for further integration with the Flavivirus host association data.
